## Billboard Hot 100

link: https://www.billboard.com/charts/hot-100

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.billboard.com/charts/hot-100")
doc = BeautifulSoup(response.text)
```

```python
content = doc.find(class_="chart-details")
songs = content.find_all(class_="chart-list-item__cursor-pointer")

rows = []

for song in songs:
    row = {}
    row['rank'] = song.find(class_="chart-list-item__rank").text.strip()
    row['artist'] = song.find(class_="chart-list-item__artist").text.strip()
    row['song'] = song.find(class_="chart-list-item__title-text").text.strip()
    rows.append(row)
    
```

```python
rows[0]
```

```python
df = pd.DataFrame(rows, columns=["rank", "artist", "song"])
df
```

```python
df.to_csv("billboard-hot-100.csv", index=False)
```
