# Epicurious (Pagination, scraping 1x per row)
Link : https://www.epicurious.com/search/cucumbers 

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
rows = []

for page_num in range(1,11):
    url = "https://www.epicurious.com/search/cucumbers?page=" + str(page_num)
    print("Now Scraping", url)
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    dishes = doc.find_all("article")
    for dish in dishes:
        row = {}
        
        try: row['tag'] = dish.find(class_="tag").text.strip()
        except: pass
        
        row['title'] = dish.find('h4').text.strip()
        
        try: row['summary'] = dish.find(class_="dek").text.strip()
        except: pass
        
        try: row['rating'] = dish.find(itemprop="ratingValue").text.strip()
        except: pass
        
        try: row['make again'] = dish.find(class_="make-again-percentage").text.strip()
        except: pass
        
        try: row['link'] = dish.find('h4').find('a')['href']
        except: pass
        
        rows.append(row)
        
rows
```

```python
links_df = pd.DataFrame(rows)
links_df
```

```python
links_df.to_csv("cucumbers.csv", index=False)
```

# Pagination

```python
def scrape_page(row):
    url = "https://www.epicurious.com" + row['link']
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    ## Print out the link column
    print("Scraping", row['link'])
    
    page = {}
    
    page['ingredients'] = []
    for ingredient in doc.find(class_="ingredient-groups"):
            page['ingredients'].append(ingredient.text.strip())
#             print(ingredient.text.strip())
    
    page['directions'] = []
    for direction in doc.find(class_="instructions"):
            page['directions'].append(direction.text.strip())
#             print(direction.text.strip())
            
    
    page['tags'] = []
    for tag in doc.find(class_="tags").find_all(itemprop="recipeCategory"):
            page['tags'].append(tag.text.strip())
#             print(tag.text.strip())
    
    return pd.Series(page)
```

```python
recipe_pages_df = links_df[links_df['tag'] == "recipe"].apply(scrape_page, axis=1)
```

```python
all_df = recipe_pages_df.merge(links_df, left_index=True, right_index=True)
all_df.head(20)
```

```python
all_df.to_csv("recipes.csv", index=False)
```
