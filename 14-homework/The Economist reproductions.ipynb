```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.patches as mpatches
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv("data/bank-profits.csv")
```

```python
colors=['#5F91A8', '#01516C', '#00A3DB', '#6DD0F7', '#018A83', '#67A8B4']

ax = df.plot(kind='pie',y='year-2007', title='2007', labels=df.region, figsize=(7,7), colors=colors)
ax.set_ylabel('')
ax.legend(bbox_to_anchor=(1.4, .6))

plt.tight_layout()
plt.savefig('eco_11.pdf')
```

```python
colors=['#5F91A8', '#01516C', '#00A3DB', '#6DD0F7', '#018A83', '#67A8B4']

ax = df.plot(kind='pie', y='year-2011', title='2011', labels=df.region, figsize=(7,7), colors=colors)
ax.set_ylabel('')
ax.legend(bbox_to_anchor=(1.4, .6))
plt.tight_layout()
plt.savefig('eco_12.pdf')
```

```python
fig, ax = plt.subplots()
ax.hlines(xmin=df['year-2007'], xmax=df['year-2011'], y=df.index, alpha=.6)
legend = [ax.scatter(x=df['year-2011'], y=df.index, s=100), ax.scatter(x=df['year-2007'], y=df.index, color='red')]


alt = ax.twinx()
# Set the new graph to have its tick marks in the same position
alt.set_yticks(ax.yaxis.get_ticklocs()) 
alt.set_ylim(ax.get_ylim())


ax.set_yticks(df.index)
ax.yaxis.set_ticklabels(df['region'])
ax.legend(legend, ['2011', '2007'])
ax.set_title('Pre-Tax Profits of the 1000 Largest Banks')
plt.tight_layout()
plt.savefig("eco_3.pdf",bbox_inches="tight", facecolor='w')
```
